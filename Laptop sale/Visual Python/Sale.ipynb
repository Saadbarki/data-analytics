```python
df.fillna(df.mode().iloc[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
average_price = df['Price_euros'].mean()
print("Average Price of Laptops:", average_price)
```

```python
most_common_brand = df['Company'].mode()[0]  # Replace 'Brand' with the actual column name if different
print("Most Common Brand:", most_common_brand)
```

```python
category_counts = df['TypeName'].value_counts()  # Replace 'Category' with the actual column name if different
print("Laptops by TypeName:\n", category_counts)
```

```python
# Plot a bar chart for the number of laptops by brand
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='skyblue')  # Replace 'category_counts' with the actual variable name
plt.title("Number of Laptops by Brand")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Price_euros'], bins=20, color='lightgreen', edgecolor='black')  # Use 'Price_euros' instead of 'Price'
plt.title("Distribution of Laptop Prices")
plt.xlabel("Price (Euros)")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the average price per brand and plot
avg_price_brand = df.groupby('Company')['Price_euros'].mean()  # Use 'Company' and 'Price_euros'
avg_price_brand.plot(kind='bar', color='salmon', figsize=(10, 6))
plt.title("Average Price by Brand")
plt.xlabel("Brand")
plt.ylabel("Average Price (Euros)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Company', y='Price_euros', palette='Set2')
plt.title('Price Distribution by Brand')
plt.xlabel('Brand')
plt.ylabel('Price (Euros)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='Company', y='Price_euros', palette='muted')
plt.title('Price Distribution by Brand (Violin Plot)')
plt.xlabel('Brand')
plt.ylabel('Price (Euros)')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.pairplot(df[['Price_euros', 'Ram', 'Weight']], diag_kind='kde')
plt.suptitle('Pairwise Relationships in the Laptop Dataset', y=1.02)
plt.show()
```
